## Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

## Univariate --> One variable data visualization

```python
# Strip plot
sns.stripplot(y='tip', data=tips, jitter=True)
plt.show()
```

```python
## Grouping with stripplot
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel("tip($)")
plt.xlabel('Day')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, jitter=False)
plt.ylabel("tip($)")
plt.xlabel('Day')
plt.show()
```

```python
# Swarm plot
sns.swarmplot(x='day', y='tip', data=tips)
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex', orient='h')
plt.show()
```

### Boxplot and violonplot

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.show()
```

```python
sns.violinplot(x='day', y='tip', data=tips)
plt.show()
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(x='day', y='tip', data=tips)
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.show()
```

```python
## Combining plots
sns.violinplot(x='day', y='tip', data=tips, inner=None, color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=True)
plt.show()

```

## Bivariate--> Two variables data visualization

```python
## Joint plot
sns.jointplot(x='total_bill', y='tip',data=tips)
plt.show()
```

```python
## Density plot
sns.jointplot(x='total_bill', y='tip',data=tips, kind='kde')
plt.show()
```

```python
## Pair plot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```
